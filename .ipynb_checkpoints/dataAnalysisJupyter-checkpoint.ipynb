```python
import csv
data ={}
keyList = []
key = 0
with open('data.csv', newline='') as csvfile:
    fileReader = csv.reader(csvfile, delimiter=',', quotechar='"')
    next(csvfile)
    next(csvfile)
    next(csvfile)
    
    for row in fileReader:
        key = int(row[0])
        keyList.append(key)
        data[key, row[2]] = row


keyList.sort()
print(keyList)

```
